# TO DO 5: หาค่ามัธยฐานและค่าเฉลี่ยของแต่ละวิชา และนำไปแสดงผลบนกราฟแท่งแบบ Multiple plot

```python
# set idx => gender
mp = {}
for i in range(len(csv_data['gender'])):
    # print(csv_data['gender'][i])
    mp[i] = csv_data['gender'][i]
```

```python
#Write code below please.
import statistics

def find_mean(score_list):
    mean_m = mean_f = cnt = 0

    for i in score_list:
        if(mp[cnt] == 'male'):mean_m+=i
        if(mp[cnt] == 'female'):mean_f+=i
        cnt+=1

    mean_m /= cnt
    mean_f /= cnt

    return (mean_m, mean_f)

def find_med(score_list):
    cnt = 0
    m = f = []

    for i in score_list:
        if(mp[cnt] == 'male'):m+=[i]
        if(mp[cnt] == 'female'):f+=[i]
        cnt+=1

    med_m = statistics.median(m)
    med_f = statistics.median(f)

    return (med_m, med_f)



mean_math_m, mean_math_f = find_mean(math_score)
mean_reading_m, mean_reading_f = find_mean(reading_score)
mean_writing_m, mean_writing_f = find_mean(writing_score)

med_math_m, med_math_f = find_med(math_score)
med_reading_m, med_reading_f = find_med(reading_score)
med_writing_m, med_writing_f = find_med(writing_score)

mean_math = math_score.mean()
mean_reading = reading_score.mean()
mean_writing = writing_score.mean()

med_math = math_score.median()
med_reading = reading_score.median()
med_writing = writing_score.median()
```

```python
N = 3
mean = (mean_math, mean_reading, mean_writing)
med = (med_math, med_reading, med_writing)
width = 0.40 
index_mean = [idx for idx in range(N)]
index_med = [idx+0.40 for idx in range(N)]
index_number = [idx+0.40 / 2 for idx in range(N)]
plt.bar(index_mean, mean, width, label='Mean')
plt.bar(index_med, med, width, label='Median')
plt.ylabel('Scores')
plt.title('Mean and Median')
plt.xticks(index_number, ('Math', 'Reading', 'Writing'))
plt.legend(loc='best')
plt.show()
```
