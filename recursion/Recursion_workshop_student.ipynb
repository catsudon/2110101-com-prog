**เราจะมีการใช้ Tree ในการดูกระบวนการทำงานของ recursive function ด้วย**

```python
from IPython.display import Image
from graphviz import Digraph
```

```python
class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None       
```

```python
def visualize_tree(tree):
    def add_nodes_edges(tree, dot=None):
        # สร้าง Digraph object
        if dot is None:
            dot = Digraph()
            dot.node(name=str(tree), label=str(tree.val))

        # เพิ่ม nodes
        if tree.left:
            dot.node(name=str(tree.left) ,label=str(tree.left.val))
            dot.edge(str(tree), str(tree.left))
            dot = add_nodes_edges(tree.left, dot=dot)
            
        if tree.right:
            dot.node(name=str(tree.right) ,label=str(tree.right.val))
            dot.edge(str(tree), str(tree.right))
            dot = add_nodes_edges(tree.right, dot=dot)

        return dot
    
    # เพิ่ม nodes วนไปเรื่อยๆ และสร้าง list ของ edges
    dot = add_nodes_edges(tree)

    # Visualize the graph
    display(dot)
```

**ตัวอย่างการใช้ TreeNode และฟังก์ชั่น visualize_tree**

```python
t1 = TreeNode(1)
t1.left = TreeNode(3)
t1.left.left = TreeNode(5)
t1.right = TreeNode(2)
```

```python
visualize_tree(t1)
```

### 1.1 สร้างฟังก์ชั่นที่ทำงานตามสมการฟังก์ชั่นด้านล่างโดยใช้เพียงแค่ for loop###

```python
# TODO 1.1 สร้างฟังก์ชั่น
def factorial(value):

    sum = 1
    for i in range(1, value+1):
        sum *= i

    return sum
```

```python
print(factorial(3))
print(factorial(4))
print(factorial(10))
```

### 1.2 สร้างฟังก์ชั่น recursive ที่ทำงานตามสมการฟังก์ชั่นด้านล่าง ###

**ตัวอย่าง:** \\
* factorial(3) --> 6 \\
* factorial(4) --> 24 \\
* factorial(10) --> 3628800

```python
# TODO 1.2.1 สร้างฟังก์ชั่น recursive
def factorial(value):
  
  if value == 1:
    return value
  
  ## เติมโค้ดด้านล่าง ##

  ##############
  
  return value*factorial(value-1)
```

```python
print(factorial(3))
print(factorial(4))
print(factorial(10))
```

```python
# TODO 1.2.2 สร้างฟังก์ชั่น recursive โดยที่ให้สร้าง tree ของฟังก์ชั่นด้วย
def factorial(value):
  
  if(value == 1):
    return 1, TreeNode(1)

  sum,newnode = factorial(value-1)
  tree = TreeNode(value*sum)
  tree.left = newnode
  return value*sum,tree
```

```python
fac,tree = factorial(3)
visualize_tree(tree)
```

### 1.3 สร้างฟังก์ชั่น recursive ที่ทำงานตามสมการฟังก์ชั่นด้านล่าง ### 

![alt text](https://i.imgur.com/uVpuc32.png)

**ตัวอย่าง:** \\
* fibonacci(4) --> 3 
* fibonacci(10) --> 55 
* fibonacci(30) --> 832040 


```python
# TODO 1.3.1 สร้างฟังก์ชั่น recursive
def fibonacci(value):
  
  ## เติมโค้ดด้านล่าง ##
  if(value == 0):
    return 0
  
  if(value == 1):
    return 1
  ##############
  
  return fibonacci(value-1) + fibonacci(value-2)  
```

```python
print(fibonacci(4))
print(fibonacci(10))
print(fibonacci(30))
```

```python
# TODO 1.3.2 สร้างฟังก์ชั่น recursive โดยที่ให้สร้าง tree ของฟังก์ชั่นด้วย
def fibonacci(value):
  
  ## เติมโค้ดด้านล่าง ##
  if(value == 0):
    return 0,TreeNode(0)
  
  if(value == 1):
    return 1,TreeNode(1)
  ##############
  
  f1,tl = fibonacci(value-1)
  f2,tr = fibonacci(value-2)
  T = TreeNode(value)
  T.left = tl
  T.right = tr
  return f1+f2, T
```

```python
fibo,tree = fibonacci(6)
visualize_tree(tree)
print(fibo)
```


### 1.4 สร้างฟังก์ชั่น recursive ที่จะสร้าง tree คล้ายรูปข้างล่าง ### 
 \\
![tree](https://drive.google.com/uc?export=view&id=1Q4_j1quw2U32GK5ho7jsUqhMbCKg-1u6)

```python
# TODO 1.4 สร้างฟังก์ชั่น recursive ที่ทำงานตาม tree ด้านบน
def create_tree(value):
  ## เติมโค้ดด้านล่าง ##
  
  if(value <= 1):
    return TreeNode(1)
  ##############
  
  tl = create_tree(value//2)
  tr = create_tree(value//2)
  T = TreeNode(value)
  T.left = tl
  T.right = tr
  return T
```

```python
bruh = create_tree(10)
visualize_tree(bruh)
```

* **การเติม: เราจะให้ฟังก์ชั่นรับ m และ n-1**
